## Question 1

For the month of July 2024, how many transactions did loyalty program members and non-members make? Compare the transaction counts between these two groups.

```python
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])

july_2024_txns = fct_transactions[
    (fct_transactions['transaction_date'].dt.year == 2024) &
    (fct_transactions['transaction_date'].dt.month == 7)
]

merged_df = july_2024_txns.merge(dim_customers, on='customer_id', how='left')

txn_counts = merged_df.groupby('is_loyalty_member')['transaction_id'].count().reset_index()

txn_counts.columns = ['is_loyalty_member', 'transaction_count']

print(txn_counts)
```

## Question 2

What is the average transaction value for loyalty program members and non-members during July 2024? Use this to identify which group has a higher average transaction value.

```python
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])

july_2024_txns = fct_transactions[
    (fct_transactions['transaction_date'].dt.year == 2024) &
    (fct_transactions['transaction_date'].dt.month == 7)
]

merged_df = july_2024_txns.merge(dim_customers, on='customer_id', how='left')

avg_txn_value = merged_df.groupby('is_loyalty_member')['transaction_value'].mean().reset_index()

avg_txn_value.columns = ['is_loyalty_member', 'avg_transaction_value']

print(avg_txn_value)

higher_group = avg_txn_value.loc[avg_txn_value['avg_transaction_value'].idxmax(), 'is_loyalty_member']
print(f"Higher average transaction value group: {'Loyalty Members' if higher_group else 'Non-Members'}")
```

## Question 3

Determine the percentage difference in average transaction value between loyalty program members and non-members for July 2024.

```python
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])

july_2024_txns = fct_transactions[
    (fct_transactions['transaction_date'].dt.year == 2024) &
    (fct_transactions['transaction_date'].dt.month == 7)
]

merged_df = july_2024_txns.merge(dim_customers, on='customer_id', how='left')

avg_txn_value = merged_df.groupby('is_loyalty_member')['transaction_value'].mean().reset_index()

loyalty_avg = avg_txn_value.loc[avg_txn_value['is_loyalty_member'] == True, 'transaction_value'].values[0]
non_loyalty_avg = avg_txn_value.loc[avg_txn_value['is_loyalty_member'] == False, 'transaction_value'].values[0]

pct_diff = ((loyalty_avg - non_loyalty_avg) / non_loyalty_avg) * 100

print(f"Percentage difference in average transaction value: {pct_diff:.2f}%")
```
