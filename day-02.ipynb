## Question 1

What is the average click-through rate (CTR) for sponsored product ads for each product category that contains the substring 'Electronics' in its name during October 2024? This analysis will help determine which electronics-related categories are performing optimally.

```python
# Note: pandas and numpy are already imported as pd and np
# The following tables are loaded as pandas DataFrames with the same names: fct_ad_performance, dim_product
# Please print your final result or dataframe

# Step 1: Convert 'recorded_date' to datetime (if not already)
fct_ad_performance['recorded_date'] = pd.to_datetime(fct_ad_performance['recorded_date'])

# Step 2: Filter for October 2024
october_ads = fct_ad_performance[
    (fct_ad_performance['recorded_date'].dt.month == 10) &
    (fct_ad_performance['recorded_date'].dt.year == 2024)
]

# Step 3: Merge with dim_product on product_id
merged_df = pd.merge(october_ads, dim_product, on='product_id', how='inner')

# Step 4: Filter for product categories containing 'Electronics'
electronics_df = merged_df[merged_df['product_category'].str.contains('Electronics', case=False, na=False)]

# Step 5: Group by product_category and calculate CTR
result = (
    electronics_df
    .groupby('product_category')
    .agg({'clicks': 'sum', 'impressions': 'sum'})
    .assign(avg_ctr=lambda x: (x['clicks'] / x['impressions']).round(4))
    .reset_index()
    .sort_values(by='avg_ctr', ascending=False)
)

print(result)
```

## Question 2

Which product categories have a CTR greater than the aggregated overall average CTR for sponsored product ads during October 2024? This analysis will identify high-performing categories for further optimization. For this question, we want to calculate CTR for each ad, then get the average across ads by product category & overall.

```python
# Step 1: Convert recorded_date to datetime (if needed)
fct_ad_performance['recorded_date'] = pd.to_datetime(fct_ad_performance['recorded_date'])

# Step 2: Filter data for October 2024
october_ads = fct_ad_performance[
    (fct_ad_performance['recorded_date'].dt.year == 2024) &
    (fct_ad_performance['recorded_date'].dt.month == 10)
]

# Step 3: Compute CTR per ad
october_ads = october_ads.copy()
october_ads['ctr'] = october_ads['clicks'] / october_ads['impressions']

# Step 4: Merge with product data
merged_df = pd.merge(october_ads, dim_product, on='product_id', how='inner')

# Step 5: Compute average CTR per product category
category_avg_ctr = (
    merged_df
    .groupby('product_category')['ctr']
    .mean()
    .reset_index()
    .rename(columns={'ctr': 'avg_ctr'})
)

# Step 6: Compute overall average CTR across all ads
overall_avg_ctr = merged_df['ctr'].mean()

# Step 7: Filter categories with avg CTR > overall CTR
high_performing_categories = category_avg_ctr[category_avg_ctr['avg_ctr'] > overall_avg_ctr]

# Step 8: Print results
print("Overall Average CTR across all sponsored ads (Oct 2024):", round(overall_avg_ctr, 4))
print("\nProduct Categories with CTR > overall average:")
print(high_performing_categories.sort_values(by='avg_ctr', ascending=False))
```

## Question 3

For the product categories identified in the previous question, what is the percentage difference between their CTR and the overall average CTR for October 2024? This analysis will quantify the performance gap to recommend specific categories for targeted advertising optimization.

```python
# Step 1: Use previous results (assuming category_avg_ctr and overall_avg_ctr already exist)

# Step 2: Filter high-performing categories again
high_performing_categories = category_avg_ctr[category_avg_ctr['avg_ctr'] > overall_avg_ctr].copy()

# Step 3: Calculate percentage difference
high_performing_categories['pct_diff_from_overall'] = (
    (high_performing_categories['avg_ctr'] - overall_avg_ctr) / overall_avg_ctr * 100
).round(2)

# Step 4: Print final result
print("Overall Average CTR (Oct 2024):", round(overall_avg_ctr, 4))
print("\nHigh-Performing Categories with % Difference from Overall CTR:")
print(high_performing_categories.sort_values(by='pct_diff_from_overall', ascending=False))
```
