## Question 1

Between April 1st and June 30th, 2025, what is the count of transactions for each payment method? This analysis will establish the baseline distribution of how customers currently pay.

```python
import pandas as pd

fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'], errors='coerce')

start_date = pd.Timestamp('2025-04-01')
end_date   = pd.Timestamp('2025-06-30 23:59:59')

mask = (fct_transactions['transaction_date'] >= start_date) & (fct_transactions['transaction_date'] <= end_date)
df_q2 = fct_transactions.loc[mask].copy()

df_q2['payment_method'] = df_q2['payment_method'].fillna('Unknown').astype(str).str.strip()

counts = (
    df_q2.groupby('payment_method', as_index=False)['transaction_id']
    .nunique()
    .rename(columns={'transaction_id':'transaction_count'})
    .sort_values('transaction_count', ascending=False)
    .reset_index(drop=True)
)

total = counts['transaction_count'].sum()
counts = pd.concat([counts, pd.DataFrame([{'payment_method':'TOTAL','transaction_count':total}])], ignore_index=True)

print(counts.to_string(index=False))
```

## Question 2

Between April 1st and June 30th, 2025, what is the average order value for each payment method? This metric will help us assess which payment methods are tied to higher spending levels.

```python
import pandas as pd

fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'], errors='coerce')

start_date = pd.Timestamp('2025-04-01')
end_date   = pd.Timestamp('2025-06-30 23:59:59')
mask = (fct_transactions['transaction_date'] >= start_date) & (fct_transactions['transaction_date'] <= end_date)
df_q2 = fct_transactions.loc[mask].copy()

df_q2['payment_method'] = df_q2['payment_method'].fillna('Unknown').astype(str).str.strip()

avg_order_values = (
    df_q2.groupby('payment_method', as_index=False)['order_value']
    .mean()
    .rename(columns={'order_value': 'avg_order_value'})
    .sort_values('avg_order_value', ascending=False)
    .reset_index(drop=True)
)

print(avg_order_values)
```

## Question 3

Between April 1st and June 30th, 2025, what would be the predicted sales lift if a 'pay over time' option were introduced? Assume that 20% of credit card transactions during this period would switch to using the 'pay over time' option. And that for these switched transactions, the order value is expected to increase by 15% based on the average order value of all credit card transactions in that same time period.

```python
import pandas as pd

fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'], errors='coerce')

start_date = pd.Timestamp('2025-04-01')
end_date   = pd.Timestamp('2025-06-30 23:59:59')
mask = (
    (fct_transactions['transaction_date'] >= start_date) &
    (fct_transactions['transaction_date'] <= end_date) &
    (fct_transactions['payment_method'].str.strip().str.lower() == 'credit card')
)
cc_df = fct_transactions.loc[mask].copy()

cc_count = cc_df['transaction_id'].nunique()
switch_count = cc_count * 0.20

cc_aov = cc_df['order_value'].mean()

increase_per_txn = cc_aov * 0.15

sales_lift = switch_count * increase_per_txn

print(f"Credit Card transactions (Q2 2025): {cc_count}")
print(f"Transactions switching (20%): {switch_count}")
print(f"Credit Card AOV: ${cc_aov:,.2f}")
print(f"Increase per switched transaction (15% uplift): ${increase_per_txn:,.2f}")
print(f"Estimated total sales lift: ${sales_lift:,.2f}")
```
